# Testing different alpha and gamma values

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
file_path = "data/alpha_gamma_results.csv"
data = pd.read_csv(file_path)

# Correct the pivot syntax
heatmap_data = data.pivot(index="alpha_init", columns="gamma_init", values="final_test_accuracy")

# Create the heatmap again with corrected data preparation
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="viridis", cbar_kws={'label': 'Final Test Accuracy'})
plt.title("Final Test Accuracy depending on Alpha and Gamma")
plt.xlabel("Gamma Init")
plt.ylabel("Alpha Init")
plt.show()
```

# Differences between ensemble sizes

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'data/ensemble_size_results.csv'
df = pd.read_csv(file_path)

# Extract the relevant columns: ensemble_size, lr, and avg_test_accuracy
df_subset = df[['ensemble_size', 'lr', 'avg_test_accuracy']]

# Create the plot
plt.figure(figsize=(10, 6))

# Create a plot for each learning rate
for lr_value in df_subset['lr'].unique():
    subset = df_subset[df_subset['lr'] == lr_value]
    plt.plot(subset['ensemble_size'], subset['avg_test_accuracy'], label=f'LR={lr_value}')

# Label the axes and add a title
plt.xlabel('Ensemble Size')
plt.ylabel('Average Test Accuracy')
plt.title('Average Test Accuracy vs. Ensemble Size for Different Learning Rates')
plt.legend(title='Learning Rate')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'data/ensemble_size_results.csv'
df = pd.read_csv(file_path)

# Extract the relevant columns: ensemble_size, lr, and avg_test_accuracy
df_subset = df[['ensemble_size', 'lr', 'final_test_accuracy']]

# Create the plot
plt.figure(figsize=(10, 6))

# Create a plot for each learning rate
for lr_value in df_subset['lr'].unique():
    subset = df_subset[df_subset['lr'] == lr_value]
    plt.plot(subset['ensemble_size'], subset['final_test_accuracy'], label=f'LR={lr_value}')

# Label the axes and add a title
plt.xlabel('Ensemble Size')
plt.ylabel('Final Test Accuracy')
plt.title('Final Test Accuracy vs. Ensemble Size for Different Learning Rates')
plt.legend(title='Learning Rate')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Pivot the data to create a heatmap-friendly format
heatmap_data = df.pivot_table(values='avg_test_accuracy', 
                              index='ensemble_size', 
                              columns='lr')

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Average Test Accuracy'})
plt.title('Heatmap of Average Test Accuracy by Ensemble Size and Learning Rate')
plt.xlabel('Learning Rate')
plt.ylabel('Ensemble Size')
plt.show()
```

```python
# If there are more numerical columns, you can use a pair plot
sns.pairplot(df[['ensemble_size', 'lr', 'avg_test_accuracy', 'batch_size']], hue='lr')
plt.title('Pair Plot of Selected Features')
plt.show()
```

```python
# Box plot of average test accuracy grouped by learning rate
plt.figure(figsize=(10, 6))
sns.boxplot(x='lr', y='avg_test_accuracy', data=df)
plt.title('Box Plot of Test Accuracy for Different Learning Rates')
plt.xlabel('Learning Rate')
plt.ylabel('Average Test Accuracy')
plt.show()
```
