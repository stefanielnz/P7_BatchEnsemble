```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm  # For progress bars

# Set random seed for reproducibility
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)

# Hyperparameters (You can adjust these)
num_epochs = 10           # Number of training epochs
batch_size = 128         # Batch size for training and validation
learning_rate = 0.001    # Learning rate for optimizer
weight_decay = 1e-4      # Weight decay (L2 regularization)
validation_split = 0.1   # Fraction of training data to use for validation
num_heads = 8            # Number of ensemble members (heads)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# CIFAR-20 Superclass Mapping
cifar_20_superclass_map = {
    0: [4, 30, 55, 72, 95],            # aquatic mammals
    1: [1, 32, 67, 73, 91],            # fish
    2: [54, 62, 70, 82, 92],           # flowers
    3: [9, 10, 16, 28, 61],            # food containers
    4: [0, 51, 53, 57, 83],            # fruit and vegetables
    5: [22, 39, 40, 86, 87],           # household electrical devices
    6: [5, 20, 25, 84, 94],            # household furniture
    7: [6, 7, 14, 18, 24],             # insects
    8: [3, 42, 43, 88, 97],            # large carnivores
    9: [12, 17, 37, 68, 76],           # large man-made outdoor things
    10: [23, 33, 49, 60, 71],          # large natural outdoor scenes
    11: [15, 19, 21, 31, 38],          # large omnivores and herbivores
    12: [34, 63, 64, 66, 75],          # medium-sized mammals
    13: [26, 45, 77, 79, 99],          # non-insect invertebrates
    14: [2, 11, 35, 46, 98],           # people
    15: [27, 29, 44, 78, 93],          # reptiles
    16: [36, 50, 65, 74, 80],          # small mammals
    17: [8, 13, 48, 58, 90],           # trees
    18: [41, 69, 81, 85, 89],          # vehicles 1
    19: [47, 52, 56, 59, 96]           # vehicles 2
}

# Function to remap CIFAR-100 labels to CIFAR-20
def remap_labels_to_cifar20(dataset):
    label_map = {}
    for superclass, subclasses in cifar_20_superclass_map.items():
        for subclass in subclasses:
            label_map[subclass] = superclass
    dataset.targets = [label_map[label] for label in dataset.targets]
    return dataset

# Data preprocessing
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.4914, 0.4822, 0.4465],  # CIFAR mean
        std=[0.2023, 0.1994, 0.2010]   # CIFAR std
    ),
])

# Load CIFAR-100 datasets
full_train_dataset = datasets.CIFAR100(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.CIFAR100(root='./data', train=False, download=True, transform=transform)

# Remap CIFAR-100 labels to CIFAR-20
full_train_dataset = remap_labels_to_cifar20(full_train_dataset)
test_dataset = remap_labels_to_cifar20(test_dataset)

# Split train dataset into training and validation
train_size = int((1 - validation_split) * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(full_train_dataset, [train_size, val_size])

# Data loaders
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
val_loader = torch.utils.data.DataLoader(dataset=val_dataset, batch_size=batch_size, shuffle=False)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

# VGG11 with Combined Parameter Sharing
class VGG11CombinedSharing(nn.Module):
    def __init__(self, num_classes=20, num_heads=4,
                 intra_share='none', inter_share='no_sharing'):
        super(VGG11CombinedSharing, self).__init__()
        self.num_heads = num_heads
        self.intra_share = intra_share
        self.inter_share = inter_share

        # Shared layers across ensemble members
        self.shared_layers_inter = nn.Sequential()

        # Shared convolutional layers within ensemble members
        self.shared_conv_64 = nn.Conv2d(64, 64, kernel_size=3, padding=1)
        self.shared_conv_256 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.shared_conv_512 = nn.Conv2d(512, 512, kernel_size=3, padding=1)

        self.heads = nn.ModuleList()

        # Define the layers of VGG11
        layers = [
            # Conv1
            ('conv1_1', nn.Conv2d(3, 64, kernel_size=3, padding=1)),
            ('relu1_1', nn.ReLU(inplace=True)),
            ('conv1_2', nn.Conv2d(64, 64, kernel_size=3, padding=1)),
            ('relu1_2', nn.ReLU(inplace=True)),
            ('pool1', nn.MaxPool2d(kernel_size=2, stride=2)),

            # Conv2
            ('conv2_1', nn.Conv2d(64, 128, kernel_size=3, padding=1)),
            ('relu2_1', nn.ReLU(inplace=True)),
            ('pool2', nn.MaxPool2d(kernel_size=2, stride=2)),

            # Conv3
            ('conv3_1', nn.Conv2d(128, 256, kernel_size=3, padding=1)),
            ('relu3_1', nn.ReLU(inplace=True)),

            # Conv4
            ('conv4_1', nn.Conv2d(256, 256, kernel_size=3, padding=1)),
            ('relu4_1', nn.ReLU(inplace=True)),
            ('pool3', nn.MaxPool2d(kernel_size=2, stride=2)),

            # Conv5
            ('conv5_1', nn.Conv2d(256, 512, kernel_size=3, padding=1)),
            ('relu5_1', nn.ReLU(inplace=True)),

            # Conv6
            ('conv6_1', nn.Conv2d(512, 512, kernel_size=3, padding=1)),
            ('relu6_1', nn.ReLU(inplace=True)),
            ('pool4', nn.MaxPool2d(kernel_size=2, stride=2)),

            # Conv7
            ('conv7_1', nn.Conv2d(512, 512, kernel_size=3, padding=1)),
            ('relu7_1', nn.ReLU(inplace=True)),

            # Conv8
            ('conv8_1', nn.Conv2d(512, 512, kernel_size=3, padding=1)),
            ('relu8_1', nn.ReLU(inplace=True)),
            ('pool5', nn.MaxPool2d(kernel_size=2, stride=2)),
        ]

        # Determine which layers to share across ensemble members (inter-model sharing)
        if self.inter_share == 'no_sharing':
            inter_shared_indices = []
        elif self.inter_share == 'early':
            inter_shared_indices = list(range(5))  # Up to 'pool1'
        elif self.inter_share == 'deep':
            inter_shared_indices = list(range(5, len(layers)))  # From 'conv2_1' onwards
        elif self.inter_share == 'all':
            inter_shared_indices = list(range(len(layers)))  # All layers
        else:
            raise ValueError("Invalid inter_share option.")

        # Build shared layers across ensemble members
        for idx in inter_shared_indices:
            name, layer = layers[idx]
            self.shared_layers_inter.add_module(name, layer)

        # Build per-head layers
        for _ in range(num_heads):
            head_layers = nn.Sequential()
            for idx, (name, layer) in enumerate(layers):
                if idx not in inter_shared_indices:
                    # Implement intra-model sharing within each head
                    if self.intra_share in ['all', 'early'] and name.startswith('conv1'):
                        # Share conv1_2 layer within the head
                        if name == 'conv1_2':
                            head_layers.add_module(name, self.shared_conv_64)
                        else:
                            head_layers.add_module(name, layer)
                    elif self.intra_share in ['all', 'deep'] and name in ['conv4_1', 'conv6_1', 'conv7_1', 'conv8_1']:
                        # Share conv layers in deep part
                        if name.startswith('conv4'):
                            head_layers.add_module(name, self.shared_conv_256)
                        else:
                            head_layers.add_module(name, self.shared_conv_512)
                    else:
                        head_layers.add_module(name, layer)
            # Add classifier
            head_layers.add_module('flatten', nn.Flatten())
            head_layers.add_module('classifier', nn.Linear(512, num_classes))
            self.heads.append(head_layers)

        # If all layers are shared across ensemble members, add classifier to shared layers
        if self.inter_share == 'all':
            self.shared_layers_inter.add_module('flatten', nn.Flatten())
            self.shared_layers_inter.add_module('classifier', nn.Linear(512, num_classes))

    def forward(self, x):
        outputs = []
        if self.inter_share == 'deep':
            # For deep inter-model sharing, apply per-head layers first
            per_head_outputs = []
            for head in self.heads:
                # Apply per-head layers up to the split point
                out = head[:5](x.clone())  # Up to 'pool1'
                per_head_outputs.append(out)
            # Apply shared layers
            shared_outs = []
            for out in per_head_outputs:
                shared_out = self.shared_layers_inter(out)
                shared_outs.append(shared_out)
            # Apply per-head classifier layers
            for idx, out in enumerate(shared_outs):
                # The classifier is the last two layers in the per-head layers
                classifier = self.heads[idx][-2:]  # 'flatten' and 'classifier'
                out = classifier(out)
                outputs.append(out)
        else:
            if len(self.shared_layers_inter) > 0:
                x = self.shared_layers_inter(x)
            for head in self.heads:
                if self.inter_share == 'all':
                    out = x
                else:
                    out = head(x.clone())
                outputs.append(out)
        outputs = torch.stack(outputs)  # Shape: [num_heads, batch_size, num_classes]
        x = outputs.mean(dim=0)  # Average over ensemble members
        return x

# Function to count parameters considering both intra-model and inter-model sharing
def count_parameters_combined(model):
    total_params = sum(p.numel() for p in model.parameters())
    shared_param_ids = set()

    # Shared parameters across ensemble members (inter-model)
    if len(model.shared_layers_inter) > 0:
        shared_param_ids.update([id(p) for p in model.shared_layers_inter.parameters()])

    # Shared parameters within ensemble members (intra-model)
    if model.intra_share in ['all', 'early']:
        shared_param_ids.update([id(p) for p in model.shared_conv_64.parameters()])
    if model.intra_share in ['all', 'deep']:
        shared_param_ids.update([id(p) for p in model.shared_conv_256.parameters()])
        shared_param_ids.update([id(p) for p in model.shared_conv_512.parameters()])

    # Unique parameters
    unique_params = set()
    for param in model.parameters():
        if id(param) not in shared_param_ids:
            unique_params.add(id(param))

    num_shared_params = sum(p.numel() for p in model.parameters() if id(p) in shared_param_ids)
    num_unique_params = sum(p.numel() for p in model.parameters() if id(p) in unique_params)

    return num_unique_params, num_shared_params

# Training function with validation and progress bars
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs):
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        # Training loop with progress bar
        with tqdm(total=len(train_loader), desc=f'Epoch {epoch+1}/{num_epochs}', unit='batch') as pbar:
            for images, labels in train_loader:
                images = images.to(device)
                labels = labels.to(device)

                # Forward pass
                outputs = model(images)
                loss = criterion(outputs, labels)

                # Backward and optimize
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

                running_loss += loss.item()

                # Calculate accuracy
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

                pbar.update(1)

        epoch_loss = running_loss / len(train_loader)
        epoch_accuracy = 100 * correct / total
        train_losses.append(epoch_loss)
        train_accuracies.append(epoch_accuracy)

        # Validation loop
        model.eval()
        val_running_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images = images.to(device)
                labels = labels.to(device)

                outputs = model(images)
                loss = criterion(outputs, labels)

                val_running_loss += loss.item()

                # Calculate accuracy
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        val_loss = val_running_loss / len(val_loader)
        val_accuracy = 100 * val_correct / val_total
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)

        print(f'Epoch [{epoch+1}/{num_epochs}], '
              f'Train Loss: {epoch_loss:.4f}, Train Acc: {epoch_accuracy:.2f}%, '
              f'Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy:.2f}%')

    return train_losses, val_losses, train_accuracies, val_accuracies

# Evaluation function on test set
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in tqdm(test_loader, desc='Testing', unit='batch'):
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    test_accuracy = 100 * correct / total
    return test_accuracy

# Plotting function with training and validation metrics
def plot_results(train_losses_list, val_losses_list, train_acc_list, val_acc_list, labels):
    num_models = len(train_losses_list)
    fig, axs = plt.subplots(2, num_models, figsize=(6 * num_models, 10))

    if num_models == 1:
        axs = [axs]

    for i in range(num_models):
        # Loss plot
        axs[0, i].plot(train_losses_list[i], label='Train Loss')
        axs[0, i].plot(val_losses_list[i], label='Val Loss')
        axs[0, i].set_xlabel('Epoch')
        axs[0, i].set_ylabel('Loss')
        axs[0, i].set_title(f'{labels[i]} Loss')
        axs[0, i].legend()

        # Accuracy plot
        axs[1, i].plot(train_acc_list[i], label='Train Acc')
        axs[1, i].plot(val_acc_list[i], label='Val Acc')
        axs[1, i].set_xlabel('Epoch')
        axs[1, i].set_ylabel('Accuracy (%)')
        axs[1, i].set_title(f'{labels[i]} Accuracy')
        axs[1, i].legend()

    plt.tight_layout()
    plt.show()

# Parameter distribution plotting function
@torch.no_grad()
def plot_parameter_distribution(model, model_label="Model"):
    params = []
    tags = []
    for name, param in model.named_parameters():
        params.append(param.cpu().flatten())
        tags += [name] * param.numel()
    params = torch.cat(params).numpy()
    tags = np.array(tags)
    df = pd.DataFrame({'Parameter Value': params, 'Parameter Name': tags})

    plt.figure(figsize=(8, 6))
    sns.kdeplot(data=df, x='Parameter Value', fill=True)
    plt.title(f'Parameter Distribution of {model_label}')
    plt.xlabel('Parameter Value')
    plt.ylabel('Density')
    plt.show()

# Main function to train and evaluate models with different sharing options
def main():
    # Define sharing options
    intra_sharing_options = ['none', 'early', 'deep', 'all']
    inter_sharing_options = ['no_sharing', 'early', 'deep', 'all']

    # You can customize the combinations you want to test
    sharing_combinations = [
        ('none', 'no_sharing'),
        ('early', 'early'),
        ('deep', 'deep'),
        ('all', 'all'),
    ]

    train_losses_list = []
    val_losses_list = []
    train_acc_list = []
    val_acc_list = []
    test_accuracies_list = []
    labels = []

    for intra_share, inter_share in sharing_combinations:
        label = f'Intra: {intra_share}, Inter: {inter_share}'
        print(f"\nTraining model with sharing options: {label}")
        model = VGG11CombinedSharing(num_classes=20, num_heads=num_heads, 
                                     intra_share=intra_share, inter_share=inter_share).to(device)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)

        # Count parameters
        num_unique_params, num_shared_params = count_parameters_combined(model)
        total_params = num_unique_params + num_shared_params
        print(f"Total parameters: {total_params}")
        print(f"Unique parameters: {num_unique_params}")
        print(f"Shared parameters: {num_shared_params}")

        train_losses, val_losses, train_accuracies, val_accuracies = train_model(
            model, train_loader, val_loader, criterion, optimizer, num_epochs)
        test_accuracy = evaluate_model(model, test_loader)
        print(f'Test Accuracy on CIFAR-100 test set: {test_accuracy:.2f}%')

        train_losses_list.append(train_losses)
        val_losses_list.append(val_losses)
        train_acc_list.append(train_accuracies)
        val_acc_list.append(val_accuracies)
        test_accuracies_list.append(test_accuracy)
        labels.append(label)

        # Plot parameter distribution
        plot_parameter_distribution(model, model_label=label)

    # Plot training results
    plot_results(train_losses_list, val_losses_list, train_acc_list, val_acc_list, labels)

    # Print test accuracies
    for i, label in enumerate(labels):
        print(f'Test Accuracy for sharing options {label}: {test_accuracies_list[i]:.2f}%')

if __name__ == "__main__":
    main()
```
